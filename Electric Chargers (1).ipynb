```python
# Charging Station
# to find all info


import requests
import pandas as pd
from datetime import date

today = date.today().strftime("%Y-%m-%d-%H-%M")
url='https://apiv2.chargehub.com/api/locationsmap?latmin=-33.043436377547444&latmax=77.52840166798826&lonmin=163.2564346359713&lonmax=-49.087315364028726&limit=1000&key=olitest&remove_networks=&remove_levels=1&remove_connectors=&remove_other=&above_power=&show_pending=1'
response=requests.get(url)
ids=[]
PortsAvail=[]

for i in response.json():
    ids.append(i['LocID'])
    PortsAvail.append(i['PortsAvail'])

 
# station data
station_id=[]
address=[]
city=[]
lat=[]
long=[]
zipcode=[]
street=[]
PriceString=[]
totalports=[]
availableports=[]
networkname=[]
AccessTime=[]
country=[]
state=[]

# plug data
plugname=[]
LastTimeUsed=[]
amp=[]
volt=[]
plug_id=[]

 
# port data
portId=[]
status=[]
port_id=[]
displayName=[]


 

for j in ids:
    url2=f'https://apiv2.chargehub.com/api/stations/details?station_id={j}&language=en'
    d=requests.get(url2)
    station_id.append(d.json()['station']['Id'])
    address.append(d.json()['station']['Address'])
    city.append(d.json()['station']['City'])
    lat.append(d.json()['station']['Lat'])
    street.append(d.json()['station']['Street'])
    zipcode.append(d.json()['station']['Zip'])
    long.append(d.json()['station']['Long'])
    availableports.append(d.json()['station']['AvailablePorts'])
    totalports.append(d.json()['station']['TotalPorts'])
    PriceString.append(d.json()['station']['PriceString'])
    networkname.append(d.json()['station']['NetworkName'])
    AccessTime.append(d.json()['station']['AccessTime'])
    country.append(d.json()['station']['Country'])
    state.append(d.json()['station']['prov_state'])

   
    for plug in d.json()['station']['PlugsArray']:
        plugname.append(plug['Name'])
        LastTimeUsed.append(plug['LastTimeUsed'])
        amp.append(plug['Amp'])
        volt.append(plug['Volt'])
        plug_id.append(j)

 

    for portname in d.json()['station']['PlugsArray'][0]['Ports']:
        portId.append(portname['portId'])
        status.append(portname['status'])
        displayName.append(portname['displayName'])
        port_id.append(j)

plug ={'Plug Name':plugname,
    'LastTimeUsed':LastTimeUsed,
    'Amp':amp,
    'Station ID':plug_id,
    'fetched date': today,
    'Volt': volt
     }


port={'Port ID':portId,
      'Display Name': displayName,
    'Status':status,
    'Station ID': port_id,
    'fetched date': today
     }

 
station={'Station ID':station_id,
         'Address':address,
         'Latitude':lat,
         'Longitude':long,
         'Zipcode': zipcode,
         'Street': street,
        'City': city,
         'State':state,
        'Country': country,
         'Access Time': AccessTime,
         'PriceString':PriceString,
         'Total Ports': totalports,
         'Available Ports': availableports,
        'fetched date': today,
         'Network Name': networkname

        }
```

```python
# export the data to loacl file

pd.DataFrame(station).to_csv(f'/Users/Zack Burnside/Desktop/collections/chargers/pepsi {today}.csv')
pd.DataFrame(plug).to_csv(f'/Users/Zack Burnside/Desktop/collections/chargers/plug {today}.csv')
pd.DataFrame(port).to_csv(f'/Users/Zack Burnside/Desktop/collections/chargers/port {today}.csv')


```
