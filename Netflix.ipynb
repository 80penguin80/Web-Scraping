```python
#Apple jobs collection
import pandas as pd
import requests
from datetime import date
today = date.today().strftime("%Y-%m-%d")

title=[]
postingDate=[]
location=[]
summary=[]
jobId=[]

 
a=0
while a<=18:
    js=requests.get(f'https://jobs.netflix.com/api/search?page={a}').json()
    a=a+1

    for i in js['records']['postings']:
        title.append(i['text'])
        summary.append(i['description'])
        postingDate.append(i['created_at'])       
        location.append(i['location'])
        jobId.append(i['external_id'])

   
dictionary={'title':title,
       'today': today,
      'location':location,
       'postingDate': postingDate,
       'summary': summary,
       'jobId': jobId
      }

pd.DataFrame(dictionary)
```

```python
def export(a):
    pd.DataFrame(a).to_csv(f'/Users/Zack Burnside/Desktop/collections/netflix/netflix_jobs {today}.csv')
export(dictionary)
```
