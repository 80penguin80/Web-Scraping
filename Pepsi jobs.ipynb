```python
import pandas as pd
import requests
from datetime import date
today = date.today().strftime("%Y-%m-%d")

 

city=[]
title=[]
qualifications=[]
latitude=[]
longitude=[]
slug=[]
create_date=[]
full_location=[]

a=1

while a<37:
    url=(f'https://www.pepsicojobs.com/api/jobs?limit=100&page={a}&sortBy=relevance&descending=false&internal=false')
    
    response=requests.get(url).json()['jobs']
    a=a+1
    
    for i in response:
        title.append(i['data']['title'])
        latitude.append(i['data']['latitude'])
        longitude.append(i['data']['longitude'])
        qualifications.append('None')
        slug.append(i['data']['slug'])
        create_date.append(i['data']['create_date'])
        #full_location.append(i['data']['full_location'])
 

dictionary={'title':title,
            'latitude':latitude,
            'longitude':longitude,
            'slug': slug,
            'create_date':create_date,
            'today':today
           }

pd.DataFrame(dictionary)
```

```python
# export the data to loacl file

def export(a):
    pd.DataFrame(a).to_csv(f'/Users/Zack Burnside/Desktop/collections/pepsi/pepsi {today}.csv')
export(dictionary)
```
