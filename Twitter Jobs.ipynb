```python
#twitter jobs collection
import requests
import pandas as pd
from datetime import date
today = date.today().strftime("%Y-%m-%d")

j=requests.get('https://careers.twitter.com/content/careers-twitter/en/roles.careers.search.json?location=&team=&offset=0&limit=15&sortBy=modified&asc=false').json()

title=[]
description=[]
location=[]
category=[]

 
for i in j['results']:
    title.append(i['title'])
    description.append(i['description'])
    location.append(i['locations'][0]['title'])
    category.append(i['teams'][0]['title'])

dictionary={'title':title,
       'description': description,
      'location':location,
       'category': category,
        'file_date': today
      }

pd.DataFrame(dictionary)


```

```python
def export(a):
    pd.DataFrame(a).to_csv(f'/Users/Zack Burnside/Desktop/collections/twitter/twitter_jobs {today}.csv')
export(dictionary)
```
