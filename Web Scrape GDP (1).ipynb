```python
#scrape the data for GDP Overall
##various import packages
from bs4 import BeautifulSoup
import lxml
import requests as request
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
#scrape the data for GDP overall
url_gdp= 'https://www.worldometers.info/gdp/#gdpyear'
page_gdp = request.get(url_gdp)
soup_gdp = BeautifulSoup(page_gdp.text, 'html')
table_data_gdp = soup_gdp.find('table', class_ = 'table table-striped table-bordered table-hover table-condensed table-list')
headers_gdp = []
for i in table_data_gdp.find_all('th'):
    title_gdp = i.text
    headers_gdp.append(title_gdp)
df_gdp = pd.DataFrame(columns = headers_gdp)
for j in table_data_gdp.find_all('tr')[1:]: #find the row data
    row_data_gdp = j.find_all('td')#assign the row data to "row_data"
    row_gdp = [tr.text for tr in row_data_gdp] #?
    length_gdp = len(df_gdp) #assign the number of rows to "length"
    df_gdp.loc[length_gdp] = row_gdp #define the number of rows of data you will print
df_gdp
```
