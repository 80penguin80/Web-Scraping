```python
# apple stores

import pandas as pd
import requests
from datetime import date

today = date.today().strftime("%Y-%m-%d")

new=requests.get('https://www.apple.com/rsp-web/autocomplete?locale=en_US').json()

store_name=[]
store_num=[]
slug=[]

for locale in new['storeListData']:
    if locale['hasStates'] == True:
        for state in locale['state']:
            for store in state['store']:
                store_name.append(store['name'])
                store_num.append(store['storeNumber'])
                slug.append(store['slug'])
                
    elif locale['hasStates'] != True:
        store_name.append(state['store'][0]['name'])
        store_num.append(store['storeNumber'])
        slug.append(store['slug'])
        
results={'slug': slug,
         'store_num': store_num,
         'store_name': store_name,
         'today': today
        }


pd.DataFrame(results)
```

```python
# export the data to loacl file

def export(a):
    pd.DataFrame(a).to_csv(f'/Users/Zack Burnside/Desktop/collections/apple_locations/apple_locations {today}.csv')
export(results)
```
