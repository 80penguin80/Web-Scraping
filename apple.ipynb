```python
#Apple jobs collection
import pandas as pd
import requests
from datetime import date
today = date.today().strftime("%Y-%m-%d")

title=[]
postingDate=[]
location=[]
summary=[]
jobId=[]

a=0

while a<=75:
    js=requests.get(f'https://jobs.apple.com/api/v1/search/page?location=united-states-USA&sort=relevance&page={a}').json()
    a=a+1

    for i in js['searchResults']:
        title.append(i['postingTitle'])
        summary.append(i['jobSummary'])
        postingDate.append(i['postingDate'])
        location.append(i['locations'][0]['name'])
        jobId.append(i['id'])

dictionary={'title':title,
       'today': today,
      'location':location,
       'postingDate': postingDate,
       'summary': summary,
       'jobId': jobId
      }

pd.DataFrame(dictionary)

 
```

```python
def export(a):
    pd.DataFrame(a).to_csv(f'/Users/Zack Burnside/Desktop/collections/apple/apple_jobs {today}.csv')
export(dictionary)
```
